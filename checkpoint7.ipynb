```python
import pandas as pd
import numpy as np

#créons notre dictionnaire: 

exam_data=  {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthieu', 'Laura', 'Kevin', 'Jonas'],
'note' : [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'tentatives' : [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualifier' : ['oui', 'non', 'oui', 'non', 'non', 'oui', 'oui', 'non', 'non', 'oui']}

#nos index: 
étiquettes = ['a', 'b', 'c', 'd', 'e','f', 'g', 'h', 'i', 'j']
 
 #créons notre data frame avec comme paramètre exam_data et index=étiquettes 
exo1 = pd.DataFrame(exam_data, index=étiquettes)
print(exo1)

# pour print les 3 premières lignes, on va utiliser head(3)
print(exo1.head(3))

#pour supprimer les lignes avec les valeurs nan on utilise dropna
print(exo1.dropna())

#pour extraire des colonnes, on utilise filter()
print(exo1.filter(items= ['name', 'note']))

#pour ajouter une nouvelle ligne k
exo1.loc['k'] = ['Suresh',15.5, 1, 'oui']

#pour supprimer, on utilise drop()
print(exo1.drop(columns='tentatives'))

#pour ajouter une nouvelle colonne on utilise df[succès]= valeur 
#La fonction "np.where()" prend en premier paramètre la condition à remplir. Le deuxième paramètre est la valeur à assigner lorsque la condition est remplie. Le troisième paramètre est la valeur assignée lorsque la condition n'a pas été remplie.
exo1['succès']= np.where(exo1['note'] > 10,1,0)

#exporter dans un fichier csv 
exo1.to_csv('my_data')
```
